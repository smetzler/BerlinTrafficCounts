```python
from os import listdir
import pandas as pd

dir_path= r"C:\Users\Simon\Downloads\Rohdaten_Berlin_1992-2019"
files = listdir(dir_path)
len(files)
```

```python
#count=False
#row=0
from tqdm import tqdm


#{'filename':f[:-4], 'knoten':knoten, 'bez':bez, 'datum':datum, 'kurzken':kurzken,
       #              'zeit':zeit, 'rich':rich, 'zaehl':cnt_dict, 'art':art, 'art_len':art_len, 'zaehl_len':zaehl_len}
#df = pd.DataFrame(columns=['filename','knoten','bez','datum','kurzken','zeit','art','art_len'])

df_dict={}
for indexf,f in tqdm(enumerate(files)): #(files[:500]):
    with open(dir_path+'\\'+f, encoding='ISO-8859-1',errors='ignore') as fi:
    #with open(dir_path+'\\'+f, encoding='utf-8') as fi:

        content = fi.readlines()

    content = [x.strip() for x in content]
    #print (indexf)
    

    ind={}
    for index,c in enumerate(content):
    
        
        
        #### HEADER ####
        if 'knoten='  in c: 
            knoten= c.replace('knoten=','')
            #df = df.append([{'filename':f[:-4]}],  ignore_index=True)
                        
        if 'bez='  in c: 
            bez= c.replace('bez=','')

        if 'datum='  in c: 
            datum= c.replace('datum=','')
            
        if 'zeit='  in c: 
            zeit= c.replace('zeit=','')
            
            zeit1=int(zeit[:2])
            zeit2=int(zeit[5:7])
            
            #print (zeit2-zeit1)
        if 'kurzken='  in c: 
            kurzken= c.replace('kurzken=','')
            #df.iloc[indexf, df.columns.get_loc('kurzken')] = kurzken
            #print (kurzken)
        if 'art='  in c: 
            art= c.replace('art=','')
            #art_len=len(art.split(','))
        if 'arm1='  in c: 
            arm1= c.replace('arm1=','')
        if 'arm2='  in c: 
            arm2= c.replace('arm2=','')
        if 'arm3='  in c: 
            arm3= c.replace('arm3=','')
        if 'arm4='  in c: 
            arm4= c.replace('arm4=','')
            
        if 'data='in c:
            #df = df.append([{'filename':f[:-4], 'knoten':knoten, 'bez':bez, 'datum':datum, 'kurzken':kurzken,
            #         'zeit':zeit, 'art':art, 'art_len':art_len}])
            df_dict[indexf]={'filename':f[:-4], 'knoten':knoten, 'bez':bez, 'datum':datum, 'kurzken':kurzken,
                     'zeit':zeit, 'art':art, 'arm1':arm1, 'arm2':arm2, 'arm3':arm3, 'arm4':arm4}#, 'art_len':art_len}
            

 

    
df= pd.DataFrame.from_dict(df_dict, orient='index')
    
df   
    
```

```python
def createDataTableHtml(df, path, titlee):

    df.to_html(path,classes = 'table table-striped table-bordered" id = "table_id')#, "utf-8")
    
    enco='ISO-8859-1'

    from bs4 import BeautifulSoup as Soup

    with open(path, 'r', encoding=enco) as f:
        webpage = f.read()#.decode('utf-8')

    soup = Soup(webpage)


    
    table = soup.find('table')
    title = soup.new_tag('title')
    title.string = titlee
    table.insert_before(title)
    
    
    style = soup.new_tag('style')
    style.string = 'table {display: none}'
    table.insert_before(style)


    script = soup.new_tag('script')
    script['src'] = "https://code.jquery.com/jquery-3.3.1.js"
    script['type'] = "text/javascript"
    table.insert_before(script)

    script = soup.new_tag('script')
    script['src'] = "https://cdn.datatables.net/1.10.19/js/jquery.dataTables.js"
    script['type'] = "text/javascript"
    script['charset'] = "utf8"
    table.insert_before(script)
    
    script = soup.new_tag('script')
    script['src'] = "https://cdn.datatables.net/1.10.20/js/dataTables.bootstrap.min.js"
    script['type'] = "text/javascript"
    table.insert_before(script)
    
    
    
    link = soup.new_tag('link')
    link['rel'] = "stylesheet"
    link['href'] = "https://maxcdn.bootstrapcdn.com/bootstrap/3.3.7/css/bootstrap.min.css"
    link['type'] = "text/css"
    table.insert_before(link)
    
    link = soup.new_tag('link')
    link['rel'] = "stylesheet"
    link['href'] = "https://cdn.datatables.net/1.10.19/css/jquery.dataTables.min.css"
    link['type'] = "text/css"
    table.insert_before(link)

    script = soup.new_tag('script')
    script.string = """$(document).ready( function () {
        $("#table_id").DataTable();
        $("#table_id").show()
    } );"""
    table.insert_before(script)

    bod=soup.new_tag("body")
    bod['style'] = "width:75%;margin-left: auto;margin-right: auto;"
    soup.table.wrap(bod)
    #soup.p.wrap(soup.new_tag("div")

    #print (soup)
    
    #add links
    for ite in soup.find_all('td'):
        if ite.string is not None:
            if ite.string.startswith('href=Roh'):
                a = soup.new_tag('a')
                a['href']=ite.string[5:]
                a.string="Rohdaten"
                ite.string=''

                #ite.td.insert(a)
                ite.append(a)
                
    #add links to charts
    for ite in soup.find_all('td'):
        if ite.string is not None:
            if ite.string.startswith('href=html_chart'):
                a = soup.new_tag('a')
                a['href']=ite.string[5:]
                a.string="Chart"
                ite.string=''

                #ite.td.insert(a)
                ite.append(a)
            

    source = soup.new_tag('a')
    source.string="Quelle: Verkehrslenkung Berlin / Rohdaten Verkehrszählungen, licence: CC BY-SA 3.0 DE"
    source['href'] = "https://fragdenstaat.de/anfrage/kfz-verkehrszahldaten-aus-berlin/"
    table.insert_after(source)
    

    with open(path, "w", encoding=enco) as outf:
        outf.write(str(soup))


#createDataTableHtml(data_fil, "table\df_test_auto.html")
```

```python
createDataTableHtml(df, 'index.html', 'Verkehrszaehldaten Berlin - counts table')
```
